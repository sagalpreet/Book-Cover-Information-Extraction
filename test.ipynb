```python
def is_valid_name(name):
    is_str = False

    if type (name) == str:
        name = [name]
        is_str = True

    res = []

    for i in name:
        res1 = list(cur.execute("select 1 where exists (select * from FN where first_name = ?)", [i]))
        res2 = list(cur.execute("select 1 where exists (select * from LN where first_name = ?)", [i]))
        res.append(res1 != [] or res2 != [])

    if (is_str): res = res[0]

    return res
```
